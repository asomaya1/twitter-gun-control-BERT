### We'll now make all text lowercase, remove links, and remove punctuation. We also remove hashtags that were used for labeling.

```python
import re
import string
import nltk
remove_hashtag = ['gunsense', 'gunsensepatriot', 'votegunsense', 'guncontrolnow',
                  'momsdemandaction', 'momsdemand', 'demandaplan', 'nowaynra', 'gunskillpeople', 'gunviolence', 'endgunviolence',
                  'gunrights', 'protect2a', 'molonlabe', 'molonlab', 'noguncontrol', 'progun',
                  'nogunregistry', 'votegunrights', 'firearmrights', 'gungrab', 'gunfriendly', '2ndamendment']

def clean_text(text): #code from https://www.kaggle.com/parulpandey/eda-and-preprocessing-for-bert
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = text.lower()
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
    tokenized_text = tokenizer.tokenize(text)
    remove_hashtags = [w for w in tokenized_text if w not in remove_hashtag]
    text = ' '.join(remove_hashtags)
    #text = re.sub(r'#\w+', '', text)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub(r'http\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    for h in remove_hashtag:
      text = re.sub(h, '', text)
      
    return text
```

```python
df['Tweets'] = df['Tweets'].apply(str).apply(lambda x: clean_text(x))
#df.iloc[7,0] = df.iloc[7,0].apply(str).apply(lambda x: clean_text(x))
```

### Now we also create a version with no stopwords either

```python
import nltk
from nltk.corpus import stopwords

def text_preprocessing(text):
    """
    Cleaning and parsing the text.

    """
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
    tokenized_text = tokenizer.tokenize(text)
    remove_stopwords = [w for w in tokenized_text if w not in stopwords.words('english')]
    combined_text = ' '.join(remove_stopwords)
    return combined_text
```

```python
df['Tweets'] = df['Tweets'].apply(str).apply(lambda x: text_preprocessing(x))
```
