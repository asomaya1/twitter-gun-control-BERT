```python
##tokenizer

training_data_file = '/content/gdrive/My Drive/Dataset/Subset2_Preprocessed_NoHashtag.csv'
vocabulary = '/content/gdrive/My Drive/Dataset/bert-base-uncased-vocab.txt'  ##30522 tokens

L = open(training_data_file,'r').readlines()
training_set = [{'sentence': L[i].split(',')[0], 'label': int(L[i].split(',')[1][:-1])} for i in range(1,len(L)) if len(L[i].split(',')) >= 2]
tokenizer = BertTokenizer(vocabulary)

###example for LSTM
print(training_set[0])
print(tokenizer.encode(training_set[0]['sentence']))
```
